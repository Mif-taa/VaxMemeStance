```python
def make_text_lookup(df, tweet_col, ocr_col=None):
    lookup = {}
    for _, r in df.iterrows():
        raw_idx = str(r["index"]).strip()
        idx = raw_idx if raw_idx.endswith(".png") else raw_idx + ".png"

        tweet = str(r.get(tweet_col, "") or "").strip()
        ocr   = str(r.get(ocr_col,   "") or "").strip() if ocr_col else ""
        tweet = "" if tweet.lower() == "nan" else tweet
        ocr   = "" if ocr.lower()   == "nan" else ocr

        if tweet and ocr:
            lookup[idx] = f"{tweet} [SEP] {ocr}"
        elif tweet:
            lookup[idx] = tweet
        elif ocr:
            lookup[idx] = ocr
        else:
            lookup[idx] = ""
    return lookup

train_text_lookup = make_text_lookup(train_text, TWEET_COL, OCR_COL)
eval_text_lookup  = make_text_lookup(eval_text,  TWEET_COL, OCR_COL)
test_text_lookup  = make_text_lookup(test_text,  TWEET_COL, OCR_COL)

for name, lookup in [("train", train_text_lookup),
                      ("eval",  eval_text_lookup),
                      ("test",  test_text_lookup)]:
    empty = sum(1 for v in lookup.values() if not v.strip())
    print(f"  {name}: {len(lookup)} entries  |  empty: {empty}")

print("\nSample train texts:")
for k, v in list(train_text_lookup.items())[:3]:
    print(f"  {k}: '{v[:100]}'")
print("✓ Text lookups ready")
```

```python
def build_model_csv(label_df, img_dir, text_lookup, out_path, is_test=False):
    rows = []
    if is_test:
        for fname in sorted(f for f in os.listdir(img_dir)
                            if f.lower().endswith((".jpg",".jpeg",".png",".webp"))):
            rows.append({
                "index":      fname,
                "image_path": os.path.join(img_dir, fname),
                "text":       text_lookup.get(fname, ""),
            })
    else:
        for _, row in label_df.iterrows():
            idx = str(row["index"])
            rows.append({
                "index":      idx,
                "image_path": os.path.join(img_dir, idx),
                "text":       text_lookup.get(idx, ""),
                "label":      int(row["label"]),
            })
    df = pd.DataFrame(rows)
    df.to_csv(out_path, index=False)
    empty = (df["text"].fillna("").str.strip() == "").sum() if not is_test else "n/a"
    print(f"  {os.path.basename(out_path)}: {len(df)} rows  |  empty text: {empty}")
    return df

train_csv_path = f"{CSV_DIR}/train.csv"
val_csv_path   = f"{CSV_DIR}/val.csv"
test_csv_path  = f"{CSV_DIR}/test.csv"

print("Building model CSVs:")
build_model_csv(train_df, TRAIN_IMGDIR, train_text_lookup, train_csv_path)
build_model_csv(eval_df,  EVAL_IMGDIR,  eval_text_lookup,  val_csv_path)
build_model_csv(None,     TEST_IMGDIR,  test_text_lookup,  test_csv_path, is_test=True)

print("\nSample train row:")
s = pd.read_csv(train_csv_path).iloc[0]
print(f"  index: {s['index']}")
print(f"  label: {s['label']}")
print(f"  text : '{str(s['text'])[:120]}'")
print("✓ CSVs ready")
```

```python
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from transformers import CLIPProcessor, CLIPModel, AutoModel, AutoTokenizer
from sklearn.metrics import classification_report, f1_score, precision_recall_fscore_support
import numpy as np, warnings, time
warnings.filterwarnings("ignore")

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device : {DEVICE}")
print(f"GPU    : {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'none'}")
```

```python
from PIL import Image
from torchvision import transforms


TRAIN_AUG = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.4),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.2, hue=0.05),
    transforms.RandomGrayscale(p=0.05),
    transforms.RandomAffine(degrees=10, translate=(0.08, 0.08), scale=(0.9, 1.1)),
    transforms.RandomPerspective(distortion_scale=0.2, p=0.3),
])

class VaxMemeDataset(Dataset):
    """Multimodal — image + text."""
    def __init__(self, csv_path, clip_processor, tokenizer,
                 max_text_len=256, is_test=False):  
        self.df             = pd.read_csv(csv_path)
        self.clip_processor = clip_processor
        self.tokenizer      = tokenizer
        self.max_text_len   = max_text_len
        self.is_test        = is_test
        self.augment        = TRAIN_AUG if not is_test else None

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        try:
            image = Image.open(str(row["image_path"])).convert("RGB")
            if self.augment: image = self.augment(image)
        except Exception:
            image = Image.new("RGB", (224, 224), (128, 128, 128))
        pv  = self.clip_processor(images=image, return_tensors="pt")["pixel_values"].squeeze(0)
        enc = self.tokenizer(
            str(row.get("text", "") or ""),
            max_length=self.max_text_len,
            padding="max_length", truncation=True, return_tensors="pt")
        item = {
            "index":          str(row["index"]),
            "pixel_values":   pv,
            "input_ids":      enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
        }
        if not self.is_test:
            item["label"] = torch.tensor(int(row["label"]), dtype=torch.long)
        return item


class TextOnlyDataset(Dataset):
    """Text only — no image loading."""
    def __init__(self, csv_path, tokenizer, max_text_len=256, is_test=False):
        self.df           = pd.read_csv(csv_path)
        self.tokenizer    = tokenizer
        self.max_text_len = max_text_len
        self.is_test      = is_test

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        row  = self.df.iloc[idx]
        text = str(row.get("text", "") or "").strip()
        text = "[empty]" if not text or text.lower() == "nan" else text
        enc  = self.tokenizer(
            text, max_length=self.max_text_len,
            padding="max_length", truncation=True, return_tensors="pt")
        item = {
            "index":          str(row["index"]),
            "input_ids":      enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
        }
        if not self.is_test:
            item["label"] = torch.tensor(int(row["label"]), dtype=torch.long)
        return item


class ImageOnlyDataset(Dataset):
    """Image only — no text."""
    def __init__(self, csv_path, clip_processor, is_test=False):
        self.df             = pd.read_csv(csv_path)
        self.clip_processor = clip_processor
        self.is_test        = is_test
        self.augment        = TRAIN_AUG if not is_test else None

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        try:
            image = Image.open(str(row["image_path"])).convert("RGB")
            if self.augment: image = self.augment(image)
        except Exception:
            image = Image.new("RGB", (224, 224), (128, 128, 128))
        pv   = self.clip_processor(images=image, return_tensors="pt")["pixel_values"].squeeze(0)
        item = {"index": str(row["index"]), "pixel_values": pv}
        if not self.is_test:
            item["label"] = torch.tensor(int(row["label"]), dtype=torch.long)
        return item

print("✓ All 3 Dataset classes defined")
```

```python
class TextOnlyModel(nn.Module):
    def __init__(self, model_name, num_classes=3, dropout=0.3, freeze=True):
        super().__init__()
        self.encoder = AutoModel.from_pretrained(model_name)
        if freeze:
            for p in self.encoder.parameters():
                p.requires_grad = False
        hidden = self.encoder.config.hidden_size

        self.classifier = nn.Sequential(
            nn.Linear(hidden, 512), nn.LayerNorm(512), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(512, 256),    nn.GELU(), nn.Dropout(dropout * 0.7),
            nn.Linear(256, num_classes)
        )

    def unfreeze_top_layers(self, n=4):
        for layer in list(self.encoder.encoder.layer)[-n:]:
            for p in layer.parameters(): p.requires_grad = True
        print(f"  Trainable: {sum(p.numel() for p in self.parameters() if p.requires_grad):,}")

    def forward(self, input_ids, attention_mask):
        out = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        cls = out.last_hidden_state[:, 0, :].float()
        return self.classifier(cls)


class ImageOnlyModel(nn.Module):
    def __init__(self, clip_name, num_classes=3, dropout=0.3, freeze=True):
        super().__init__()
        self.clip = CLIPModel.from_pretrained(clip_name)
        if freeze:
            for p in self.clip.parameters(): p.requires_grad = False
        clip_dim = self.clip.config.vision_config.hidden_size
        self.classifier = nn.Sequential(
            nn.Linear(clip_dim, 512), nn.LayerNorm(512), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(512, 256), nn.GELU(), nn.Dropout(dropout * 0.7),
            nn.Linear(256, num_classes)
        )

    def unfreeze_top_layers(self, n=4):
        for layer in list(self.clip.vision_model.encoder.layers)[-n:]:
            for p in layer.parameters(): p.requires_grad = True
        print(f"  Trainable: {sum(p.numel() for p in self.parameters() if p.requires_grad):,}")

    def forward(self, pixel_values):
        cls = self.clip.vision_model(
            pixel_values=pixel_values).last_hidden_state[:, 0, :]
        return self.classifier(cls)


class CrossModalAttentionLayer(nn.Module):
    def __init__(self, dim=512, num_heads=8, dropout=0.1):
        super().__init__()
        self.img2txt  = nn.MultiheadAttention(dim, num_heads, dropout=dropout, batch_first=True)
        self.txt2img  = nn.MultiheadAttention(dim, num_heads, dropout=dropout, batch_first=True)
        self.norm_img = nn.LayerNorm(dim)
        self.norm_txt = nn.LayerNorm(dim)
        self.drop     = nn.Dropout(dropout)

    def forward(self, img, txt):
        img_ctx, _ = self.img2txt(img, txt, txt)
        img = self.norm_img(img + self.drop(img_ctx))
        txt_ctx, _ = self.txt2img(txt, img, img)
        txt = self.norm_txt(txt + self.drop(txt_ctx))
        return img, txt


class VaxMemeFusionModel(nn.Module):
    def __init__(self, clip_name, text_name, fusion_dim=512,
                 num_attn_layers=2, dropout=0.3, freeze_encoders=True):
        super().__init__()
        self.clip    = CLIPModel.from_pretrained(clip_name)
        self.roberta = AutoModel.from_pretrained(text_name)
        if freeze_encoders:
            for p in list(self.clip.parameters()) + list(self.roberta.parameters()):
                p.requires_grad = False
        clip_dim    = self.clip.config.vision_config.hidden_size
        roberta_dim = self.roberta.config.hidden_size
        self.img_proj = nn.Sequential(
            nn.Linear(clip_dim, fusion_dim), nn.LayerNorm(fusion_dim), nn.GELU())
        self.txt_proj = nn.Sequential(
            nn.Linear(roberta_dim, fusion_dim), nn.LayerNorm(fusion_dim), nn.GELU())
        self.cross_attn = nn.ModuleList([
            CrossModalAttentionLayer(fusion_dim, num_heads=8, dropout=dropout)
            for _ in range(num_attn_layers)
        ])
        # IMPROVEMENT 4: LayerNorm in fusion head
        self.classifier = nn.Sequential(
            nn.Linear(5 * fusion_dim, 512), nn.LayerNorm(512), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(512, 128),            nn.GELU(), nn.Dropout(dropout * 0.7),
            nn.Linear(128, NUM_CLASSES),
        )

    def unfreeze_top_layers(self, clip_layers=4, roberta_layers=4):
        for layer in list(self.clip.vision_model.encoder.layers)[-clip_layers:]:
            for p in layer.parameters(): p.requires_grad = True
        for layer in list(self.roberta.encoder.layer)[-roberta_layers:]:
            for p in layer.parameters(): p.requires_grad = True
        print(f"  Trainable: {sum(p.numel() for p in self.parameters() if p.requires_grad):,}")

    def forward(self, pixel_values, input_ids, attention_mask):
        img = self.img_proj(
            self.clip.vision_model(pixel_values=pixel_values).last_hidden_state)
        txt = self.txt_proj(
            self.roberta(input_ids=input_ids,
                         attention_mask=attention_mask).last_hidden_state.float())
        for layer in self.cross_attn:
            img, txt = layer(img, txt)
        img_cls, txt_cls = img[:,0,:], txt[:,0,:]
        fused = torch.cat([img_cls, txt_cls, img.mean(1), txt.mean(1),
                           (img_cls - txt_cls).abs()], dim=-1)
        return self.classifier(fused)

print("✓ All 3 Model classes defined")
```

```python
from tqdm.auto import tqdm
from transformers import get_cosine_schedule_with_warmup
import time

class LabelSmoothingLoss(nn.Module):
    def __init__(self, classes=NUM_CLASSES, smoothing=0.1):
        super().__init__()
        self.smoothing, self.cls = smoothing, classes
    def forward(self, logits, targets):
        confidence = 1.0 - self.smoothing
        smooth_val = self.smoothing / (self.cls - 1)
        with torch.no_grad():
            st = torch.full_like(logits, smooth_val)
            st.scatter_(1, targets.unsqueeze(1), confidence)
        return -(st * F.log_softmax(logits, -1)).sum(-1).mean()

class WeightedLabelSmoothingLoss(nn.Module):
    def __init__(self, class_weights, classes=NUM_CLASSES, smoothing=0.1):
        super().__init__()
        self.smoothing = smoothing
        self.cls       = classes
        self.register_buffer("weights", class_weights)
    def forward(self, logits, targets):
        confidence = 1.0 - self.smoothing
        smooth_val = self.smoothing / (self.cls - 1)
        with torch.no_grad():
            st = torch.full_like(logits, smooth_val)
            st.scatter_(1, targets.unsqueeze(1), confidence)
        log_prob      = F.log_softmax(logits, -1)
        sample_loss   = -(st * log_prob).sum(-1)
        sample_weight = self.weights[targets]
        return (sample_loss * sample_weight).mean()

# ── RoBERTa supports AMP — enable for all 3 modes ────────────
scalers = {
    "text":   torch.amp.GradScaler('cuda'),   # RoBERTa: AMP safe ✓
    "image":  torch.amp.GradScaler('cuda'),
    "fusion": torch.amp.GradScaler('cuda'),
}

def train_one_epoch(model, loader, opt, sched, criterion, mode="fusion"):
    model.train()
    scaler = scalers[mode]
    total_loss, all_preds, all_labels = 0.0, [], []
    for batch in tqdm(loader, desc="  Train", leave=False):
        labels = batch["label"].to(DEVICE)
        opt.zero_grad()

        if mode == "text":
            with torch.amp.autocast('cuda'):
                logits = model(batch["input_ids"].to(DEVICE),
                               batch["attention_mask"].to(DEVICE))
                loss   = criterion(logits, labels)
            if torch.isnan(loss) or torch.isinf(loss):
                opt.zero_grad(); continue
            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(opt); scaler.update()

        elif mode == "image":
            with torch.amp.autocast('cuda'):
                logits = model(batch["pixel_values"].to(DEVICE))
                loss   = criterion(logits, labels)
            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(opt); scaler.update()

        else:  # fusion
            with torch.amp.autocast('cuda'):
                logits = model(batch["pixel_values"].to(DEVICE),
                               batch["input_ids"].to(DEVICE),
                               batch["attention_mask"].to(DEVICE))
                loss   = criterion(logits, labels)
            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(opt); scaler.update()

        sched.step()
        total_loss += loss.item()
        all_preds.extend(logits.argmax(-1).cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
    return total_loss/len(loader), f1_score(
        all_labels, all_preds, average="macro", zero_division=0)


@torch.no_grad()
def evaluate(model, loader, mode="fusion"):
    model.eval()
    all_preds, all_labels = [], []
    for batch in tqdm(loader, desc="  Val  ", leave=False):
        with torch.amp.autocast('cuda'):
            if mode == "text":
                logits = model(batch["input_ids"].to(DEVICE),
                               batch["attention_mask"].to(DEVICE))
            elif mode == "image":
                logits = model(batch["pixel_values"].to(DEVICE))
            else:
                logits = model(batch["pixel_values"].to(DEVICE),
                               batch["input_ids"].to(DEVICE),
                               batch["attention_mask"].to(DEVICE))
        all_preds.extend(logits.argmax(-1).cpu().tolist())
        all_labels.extend(batch["label"].tolist())
    macro_f1 = f1_score(all_labels, all_preds, average="macro", zero_division=0)
    report   = classification_report(
        all_labels, all_preds, target_names=LABEL_NAMES, zero_division=0)
    return macro_f1, report


def run_training(model, train_loader, val_loader, cfg, mode="fusion", cw_tensor=None):
    criterion = (WeightedLabelSmoothingLoss(cw_tensor)
                 if cw_tensor is not None else LabelSmoothingLoss())

    
    if cfg.get("unfreeze_epoch", 999) == 0:
        enc_params, head_params = [], []
        for name, p in model.named_parameters():
            if not p.requires_grad: continue
            is_enc = any(k in name for k in ["encoder", "clip", "roberta"])
            (enc_params if is_enc else head_params).append(p)
        optimizer = torch.optim.AdamW([
            {"params": enc_params,  "lr": 2e-6},
            {"params": head_params, "lr": cfg["lr"]},
        ], weight_decay=cfg["weight_decay"])
    else:
        params    = filter(lambda p: p.requires_grad, model.parameters())
        optimizer = torch.optim.AdamW(params, lr=cfg["lr"],
                                      weight_decay=cfg["weight_decay"])

    steps     = len(train_loader) * cfg["epochs"]
    scheduler = get_cosine_schedule_with_warmup(optimizer, int(0.1*steps), steps)

    best_f1, history = 0.0, []
    print(f"\n{'Ep':>3}  {'Loss':>7}  {'TrainF1':>8}  {'ValF1':>7}  {'Best':>5}  {'Sec':>5}")
    print("─" * 46)

    for epoch in range(1, cfg["epochs"] + 1):
        t0 = time.time()

        if epoch == cfg.get("unfreeze_epoch", 999):
            print(f"\n  ── Epoch {epoch}: unfreezing ──")
            if mode == "fusion":
                model.unfreeze_top_layers(4, 4)
            else:
                model.unfreeze_top_layers(4)

            enc_params, head_params = [], []
            for name, p in model.named_parameters():
                if not p.requires_grad: continue
                is_enc = any(k in name for k in ["clip", "roberta", "encoder"])
                (enc_params if is_enc else head_params).append(p)

            enc_lr = 2e-6
            optimizer = torch.optim.AdamW([
                {"params": enc_params,  "lr": enc_lr},
                {"params": head_params, "lr": cfg["lr"]},
            ], weight_decay=cfg["weight_decay"])
            remaining = len(train_loader) * (cfg["epochs"] - epoch + 1)
            scheduler = get_cosine_schedule_with_warmup(
                optimizer, int(0.1*remaining), remaining)

            
            scalers["image"]  = torch.amp.GradScaler('cuda')
            scalers["fusion"] = torch.amp.GradScaler('cuda')
            scalers["text"]   = torch.amp.GradScaler('cuda')

        loss, tr_f1    = train_one_epoch(
            model, train_loader, optimizer, scheduler, criterion, mode)
        val_f1, report = evaluate(model, val_loader, mode)
        elapsed        = time.time() - t0
        is_best        = val_f1 > best_f1

        if is_best:
            best_f1 = val_f1
            torch.save(model.state_dict(), cfg["save_path"])

        history.append({"epoch":epoch,"loss":loss,"train_f1":tr_f1,"val_f1":val_f1})
        print(f"  {epoch:2d}   {loss:7.4f}   {tr_f1:8.4f}   {val_f1:7.4f}"
              f"   {'★' if is_best else ' '}    {elapsed:4.0f}s")
        if is_best: print(f"\n{report}\n")

    print(f"\n✓ Best Val Macro-F1: {best_f1:.4f} → {cfg['save_path']}")
    return best_f1, history

print("✓ Training functions defined (AMP enabled for all modes)")
```

```python
import numpy as np
from sklearn.metrics import f1_score, precision_recall_fscore_support, classification_report

@torch.no_grad()
def get_softmax_probs(model, loader, mode="fusion"):
    model.eval()
    all_probs, all_labels, all_indices = [], [], []
    for batch in tqdm(loader, desc=f"  Probs/{mode}", leave=False):
        with torch.amp.autocast('cuda'):
            if mode == "text":
                logits = model(batch["input_ids"].to(DEVICE),
                               batch["attention_mask"].to(DEVICE))
            elif mode == "image":
                logits = model(batch["pixel_values"].to(DEVICE))
            else:
                logits = model(batch["pixel_values"].to(DEVICE),
                               batch["input_ids"].to(DEVICE),
                               batch["attention_mask"].to(DEVICE))
        probs = torch.softmax(logits, dim=-1).cpu().numpy()
        all_probs.append(probs)
        # labels only exist for train/val, not test
        if "label" in batch:
            all_labels.extend(batch["label"].tolist())
        if "index" in batch:
            all_indices.extend(batch["index"])
    return np.concatenate(all_probs, axis=0), all_labels, all_indices

print("✓ get_softmax_probs defined (handles test set)")
```

```python
text_model.load_state_dict(torch.load(TEXT_CFG["save_path"], map_location=DEVICE))
img_model.load_state_dict(torch.load(IMG_CFG["save_path"], map_location=DEVICE))
fusion_model.load_state_dict(torch.load(FUSION_CFG["save_path"], map_location=DEVICE))

print("Computing softmax probabilities for all models...")
probs_text,   true_labels, _ = get_softmax_probs(text_model,   val_text_loader,   "text")
probs_image,  _,           _ = get_softmax_probs(img_model,    val_img_loader,    "image")
probs_fusion, _,           _ = get_softmax_probs(fusion_model, val_fusion_loader, "fusion")

true_labels = np.array(true_labels)
f1_txt = f1_score(true_labels, probs_text.argmax(1),   average="macro", zero_division=0)
f1_img = f1_score(true_labels, probs_image.argmax(1),  average="macro", zero_division=0)
f1_fus = f1_score(true_labels, probs_fusion.argmax(1), average="macro", zero_division=0)

p1, r1, _, _ = precision_recall_fscore_support(true_labels, probs_text.argmax(1),   average="macro", zero_division=0)
p2, r2, _, _ = precision_recall_fscore_support(true_labels, probs_image.argmax(1),  average="macro", zero_division=0)
p3, r3, _, _ = precision_recall_fscore_support(true_labels, probs_fusion.argmax(1), average="macro", zero_division=0)

print("\n" + "="*62)
print(f"{'Model':<20} {'Classifier':<22} {'P':>6} {'R':>6} {'F1':>6}")
print("="*62)
print(f"{'Unimodal (Text)':<20} {'RoBERTa-large':<22} {p1:>6.4f} {r1:>6.4f} {f1_txt:>6.4f}")
print(f"{'Unimodal (Image)':<20} {'CLIP ViT-L/14':<22} {p2:>6.4f} {r2:>6.4f} {f1_img:>6.4f}")
print(f"{'Multimodal':<20} {'CLIP-L + RoBERTa-L':<22} {p3:>6.4f} {r3:>6.4f} {f1_fus:>6.4f}")
print("="*62)

total = f1_txt + f1_img + f1_fus
w_t, w_i, w_f = f1_txt/total, f1_img/total, f1_fus/total
print(f"\nEnsemble weights → text:{w_t:.3f}  image:{w_i:.3f}  fusion:{w_f:.3f}")

ensemble_probs = w_t * probs_text + w_i * probs_image + w_f * probs_fusion
ensemble_preds = ensemble_probs.argmax(1)
f1_ens = f1_score(true_labels, ensemble_preds, average="macro", zero_division=0)
print(f"\n{'='*62}")
print(f"Weighted Ensemble Macro-F1: {f1_ens:.4f}")
print("="*62)
print(classification_report(true_labels, ensemble_preds, target_names=LABEL_NAMES, zero_division=0))

from itertools import product as iproduct

def apply_thresholds(probs, thresholds):
    return (probs / np.array(thresholds)).argmax(1)

best_thresh_f1  = f1_ens
best_thresholds = [1.0, 1.0, 1.0]
threshold_vals  = [0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3]

print("\nSearching best per-class thresholds...")
for t0, t1, t2 in iproduct(threshold_vals, threshold_vals, threshold_vals):
    preds = apply_thresholds(ensemble_probs, [t0, t1, t2])
    f1    = f1_score(true_labels, preds, average="macro", zero_division=0)
    if f1 > best_thresh_f1:
        best_thresh_f1  = f1
        best_thresholds = [t0, t1, t2]

print(f"Best thresholds : {best_thresholds}")
print(f"Threshold-tuned Ensemble F1: {best_thresh_f1:.4f}  (raw ensemble: {f1_ens:.4f})")
print(classification_report(
    true_labels, apply_thresholds(ensemble_probs, best_thresholds),
    target_names=LABEL_NAMES, zero_division=0))
```
